# <center> Dictionary (Sözlük)

- Sırasız ver tutar
- Günlük hayatta sözlüklerile benzerlik gösterir
- {} arasında tanımlanır
- key: anahtar - bir bilgiye ulaşmak için kullanılır
- value: değer
- key value - ÖRNEĞİ: {06:"Ankara"} --> 06 key, 'Ankara' value
- dict("") şeklinde de tanımlama yapıbilir

```python
dictionary = {"book" : "kitap", "table" : "masa"}
print(type(dictionary))
```

```python
print(dictionary)
print(dictionary["table"])
print(dictionary["kitap"])
```

#### Veri Ekleme

```python
dictionary["pencil"] = "kurşun kalem"
print(dictionary)
```

#### Veri Düzenleme

```python
dictionary["pencil"] = "kalem"
print(dictionary)
```

#### Veri Silme

```python
del(dictionary["pencil"])
print(dictionary)
```

#### Value'nun sözlük olması

```python
dictValue = {1 : {"TL",  "Türkiye"}}
print(dictValue)
```

### Dictionary Methods (Sözlük metotları)

- keys(): <br>
returns keys in the dictionary
- values(): <br>
returns values in the dictionary
- items(): <br>
returns keys and values pairs in the dictionary


```python
print(dictionary.keys())
print(dictionary.values())
print(dictionary.items())
```
