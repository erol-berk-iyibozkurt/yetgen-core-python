# Problem 3
Vehicle sınıfının tüm değişkenlerini ve yöntemlerini devralacak bir alt sınıf olarak Bus sınıfı oluşturunuz.



```
Output
> Name: School Volvo Speed: 180 Mileage: 12
```


````python
class Vehicle():
    class Bus():
        def __init__(self,model,speed,mileage):
            self.model = model
            self.speed = speed
            self.mileage = mileage
    def __init__(self):
        self.bus1 = self.Bus("Fiat Ducato", 180, 12)
        self.bus2 = self.Bus("Iveco Daily", 190, 14)
        self.bus3 = self.Bus("Mercedes Sprinter", 200, 18)

    def get_buses(self):
        return [self.bus1, self.bus2, self.bus3]

vehicle = Vehicle()
buses = vehicle.get_buses()

for bus in buses:
    print(f'The model: "{bus.model}". The speed: {bus.speed} and the mileage: {bus.mileage}.')
````
