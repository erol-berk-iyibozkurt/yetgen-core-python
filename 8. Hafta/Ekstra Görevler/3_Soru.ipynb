Çeşitli metin dosyalarından karakterleri ayıklamak ve bir listeye koymak için bir Python programı yazın. ```test.txt``` ve ```demo.txt``` dosyalarından faydalanabilirsiniz.

````python
with open("test.txt", "r",encoding= "utf-8")as fi:
    read1 = fi.readlines()

with open("demo.txt", "r",encoding= "utf-8")as fi:
    read2 = fi.readlines()

l = []
totalList = read1 + read2

for i in totalList:
    for a in i:
        if a != "\n":
            l.append(a)
print(l)
````
