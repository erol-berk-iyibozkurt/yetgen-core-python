# Problem 1
Vehicle sınıfından miras alan bir Bus sınıfı oluşturun. Bus.seating_capacity()'nin kapasite bağımsız değişkenine varsayılan olarak 50 değerini verin.

```
Output:
> The seating capacity of a bus is 50 passengers
```


````python
class Vehicle:
    def __init__(self, type_of_vehicle, seating_capacity):
        self.type_of_vehicle = type_of_vehicle
        self.seating_capacity = seating_capacity

class Bus(Vehicle):
    def __init__(self):
        super().__init__('Bus', 50)

    def __str__(self):
        return f"The seating capacity of a bus is {self.seating_capacity} passengers"

# Create an instance of Bus
bus = Bus()
print(bus)
````

# Problem 2


School_bus'ın aynı zamanda Vehicle sınıfının bir örneği olup olmadığını belirleyiniz.

````python
class Vehicle:
    def __init__(self, type_of_vehicle, seating_capacity):
        self.type_of_vehicle = type_of_vehicle
        self.seating_capacity = seating_capacity

class Bus(Vehicle):
    def __init__(self):
        super().__init__('Bus', 50)

    def __str__(self):
        return f"The seating capacity of a bus is {self.seating_capacity} passengers"

# School_bus created as a item
school_bus = Bus()

# controlling school_bus in Vehichle
is_instance_of_vehicle = isinstance(school_bus, Vehicle)

print(f" Is School_bus a example for the class of Vehicle? {is_instance_of_vehicle}")
````
