### The final code:

```python
import time
import random

channel_l = {}

# File reading operation in a try-except block
try:
    with open("channel_list.txt", "r") as f:
        for line in f:
            name, number = line.strip().split()
            channel_l[name] = int(number)
except FileNotFoundError:
    print("channel_list.txt file not found. Starting with an empty channel list.")

class Controller:
    def __init__(self, TV_status="close", sound_volume=0, currently_channel=None):
        self.TV_status = TV_status
        self.sound_volume = sound_volume
        self.currently_channel = currently_channel

    def open_TV(self):
        if self.TV_status == "close":
            print("TV is opening...")
            time.sleep(3)
            self.TV_status = "open"
            print("TV is opened.")
        else:
            print("TV is already open.")

    def close_TV(self):
        if self.TV_status == "open":
            print("TV is closing...")
            time.sleep(3)
            self.TV_status = "close"
            print("TV is closed.")
        else:
            print("TV is already closed.")

    def manage_sound(self):
        if self.TV_status == "close":
            print("You need to open the TV first.")
            return

        while True:
            print(f"Current volume: {self.sound_volume}")
            choice = input("Enter '+' to increase, '-' to decrease, or 'q' to quit: ")
            if choice == '+' and self.sound_volume < 100:
                self.sound_volume = min(100, self.sound_volume + 5)
                print(f"Volume increased to {self.sound_volume}")
            elif choice == '-' and self.sound_volume > 0:
                self.sound_volume = max(0, self.sound_volume - 5)
                print(f"Volume decreased to {self.sound_volume}")
            elif choice.lower() == 'q':
                break
            else:
                print("Invalid input!")

    def add_channel(self):
        channel = input("Name of the new channel: ")
        channel_number = int(input(f"In which channel number {channel} should be moving? "))
        
        if channel_number in channel_l.values():
            print(f"You already have a channel called {list(channel_l.keys())[list(channel_l.values()).index(channel_number)]} in that number.")
            decision = input("Are you sure to change the channel number? (Y/N): ").lower()
            if decision == 'y':
                old_channel = list(channel_l.keys())[list(channel_l.values()).index(channel_number)]
                channel_l[old_channel] = max(channel_l.values()) + 1
        
        channel_l[channel] = channel_number
        print(f"Channel {channel} has been added to number {channel_number}.")

    def number_of_channels(self):
        total_channel = len(channel_l)
        if total_channel > 1:
            print(f"Your TV has {total_channel} channels.")
        else:
            print(f"Your TV has {total_channel} channel.")

    def open_channel(self):
        if self.TV_status == "close":
            print("You need to open the TV first.")
            return

        desired_channel = input("Which channel do you want to watch now? ")
        if desired_channel.isdigit():
            channel_number = int(desired_channel)
            if channel_number in channel_l.values():
                channel_name = list(channel_l.keys())[list(channel_l.values()).index(channel_number)]
                print(f"{channel_name} is opening...")
                time.sleep(2)
                self.currently_channel = channel_name
            else:
                print("There is no channel at this number.")
        elif desired_channel in channel_l:
            print(f"{desired_channel} is opening...")
            time.sleep(2)
            self.currently_channel = desired_channel
        else:
            print("This channel doesn't exist.")

    def randomize_channel(self):
        if self.TV_status == "close":
            print("You need to open the TV first.")
            return

        if not channel_l:
            print("There are no channels available.")
            return

        random_channel = random.choice(list(channel_l.keys()))
        print(f"You are being directed to the {random_channel}")
        time.sleep(2)
        self.currently_channel = random_channel

    def info_TV(self):
        print("Your TV is using the latest version.")
        if self.TV_status == "open":
            print(f"Current channel: {self.currently_channel}")
            print(f"Volume: {self.sound_volume}")

def display_menu():
    print("""
1. Open TV
2. Close TV
3. Manage Sound
4. Add Channel
5. Learn the number of channels 
6. Go to a random channel
7. Learn the info of the TV
8. Go to the desired channel
q. Quit
    """)

def main():
    controller = Controller()
    
    while True:
        display_menu()
        choice = input("Please enter an option: ")
        
        if choice == '1':
            controller.open_TV()
        elif choice == '2':
            controller.close_TV()
        elif choice == '3':
            controller.manage_sound()
        elif choice == '4':
            controller.add_channel()
        elif choice == '5':
            controller.number_of_channels()
        elif choice == '6':
            controller.randomize_channel()
        elif choice == '7':
            controller.info_TV()
        elif choice == '8':
            controller.open_channel()
        elif choice.lower() == 'q':
            print("Exiting the program...")
            break
        else:
            print("Invalid option! Please try again.")

if __name__ == "__main__":
    main()
```
